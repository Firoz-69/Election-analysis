```python
import requests
from bs4 import BeautifulSoup

# Function to scrape data from a webpage
def scrape_webpage(url):
    # Sending a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parsing the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Example: Extracting all links from the page
        links = soup.find_all('a', href=True)
        for link in links:
            print(link['href'])

        # Example: Extracting data from a table
        table = soup.find('table', class_='table-class')  # Replace with actual class or ID of the table
        if table:
            rows = table.find_all('tr')
            for row in rows:
                # Assuming each row has two cells (for example)
                cells = row.find_all('td')
                if len(cells) == 2:
                    data1 = cells[0].text.strip()
                    data2 = cells[1].text.strip()
                    print(f"{data1}: {data2}")

        # Example: Extracting data from a list
        ul_list = soup.find('ul', class_='list-class')  # Replace with actual class or ID of the list
        if ul_list:
            items = ul_list.find_all('li')
            for item in items:
                print(item.text.strip())

    else:
        print(f"Failed to retrieve webpage: {response.status_code}")

# URL of the webpage to scrape
url_to_scrape = 'https://results.eci.gov.in/'  # Replace with actual URL

# Call the function with the URL
scrape_webpage(url_to_scrape)
```
